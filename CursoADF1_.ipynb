```python
bpa_weg_varios_anos = pd.DataFrame()
pasta_drive = '/content/drive/MyDrive/2_ciencia_de_dados/3_Phyton/4_Contabilidade/geral/cursolivro'
for ano in range(2021,2024):
  # # Ler informações de arquivo (.csv) específico
  bpa = pd.read_csv(pasta_drive+f'/datasets/file_name/dfp_cia_aberta_BPA_con_{ano}.csv', sep=';', encoding='ISO-8859-9', decimal=',')


 # Filtrar informações da coluna ORDEM_EXERC == ÚLTIMO
  bpa = bpa[bpa['ORDEM_EXERC'] == 'ÚLTIMO']
  bpa.head()

  # Filtrar empresa por CNPJ
  bpa_weg = bpa[ bpa['CNPJ_CIA'] == '84.429.695/0001-11']

  # Filtrar apenas as colunas que serão utilizadas
  bpa_weg = bpa_weg[['DT_REFER','CD_CONTA','DS_CONTA','VL_CONTA']]

  # Converter valores numéricos
  bpa_weg['VL_CONTA'] = pd.to_numeric(bpa_weg['VL_CONTA'])

  # Converter datas
  bpa_weg['DT_REFER'] = pd.to_datetime(bpa_weg['DT_REFER'])

  # Criar coluna ANO
  bpa_weg['ANO'] = bpa_weg['DT_REFER'].dt.year

  # Reduzir as contas
  bpa_weg = bpa_weg[bpa_weg['CD_CONTA'].str.len() <= 7]

  # Reiniciar os indices
  bpa_weg.reset_index(inplace=True, drop=True)

  # Agrupamento em linhas
  bpa_weg_varios_anos = pd.concat([bpa_weg_varios_anos, bpa_weg])

bpa_weg_varios_anos = bpa_weg_varios_anos.pivot(index=['CD_CONTA','DS_CONTA'], columns='ANO', values='VL_CONTA')
bpa_weg_varios_anos
```

```python
#Análise Horizontal
```

```python
bpa_weg_varios_anos['2021-2022'] = (bpa_weg_varios_anos[2022] / bpa_weg_varios_anos[2021] - 1) * 100
```

```python
bpa_weg_varios_anos['2022-2023'] = (bpa_weg_varios_anos[2023] / bpa_weg_varios_anos[2022] - 1) * 100
```
