## Liquidez Imediata

```python
bpp_empresa=bpp
bpa_empresa=bpa
dre_empresa=dre
```

```python
caixa = bpa_empresa[bpa_empresa['CD_CONTA'] == '1.01.01'].loc[2,'VL_CONTA']
passivo_circulante = bpp_empresa[bpp_empresa['CD_CONTA'] == '2.01'].loc[1,'VL_CONTA']
liq_imediata = round(caixa/passivo_circulante,2)
print(f'Liquidez Imediata: {liq_imediata}')
```

## Liquidez Corrente

```python
ativo_circulante = bpa_empresa[bpa_empresa['CD_CONTA'] == '1.01'].loc[1,'VL_CONTA']
passivo_circulante = bpp_empresa[bpp_empresa['CD_CONTA'] == '2.01'].loc[1,'VL_CONTA']
liq_corrente = round(ativo_circulante/passivo_circulante,2)
print(f'Liquidez Corrente: {liq_corrente}')
```
